```python
#Feature Engineering
read_file['family_size']=read_file['Parch']+read_file['SibSp']
```

```python
read_file.sample(5)
```

```python
#now we willl engineer a new feature by the name of family type

def family_type(number):
  if number==0:
    return 'Alone'
  elif number>0 and number<=4:
    return 'Medium'
  else:
    return 'Large'
```

```python
read_file['family_type']=read_file['family_size'].apply(family_type)
```

```python
read_file.sample(5)
```

```python
#Dropping SibSp,Parch and family_size
read_file.drop(columns=['SibSp','Parch','family_size'],inplace=True)
```

```python
read_file.sample(5)
```

```python
pd.crosstab(read_file['family_type'],read_file['Survived']).apply(lambda r:round((r/r.sum())*100,1),axis=1)
```

```python
#handling outliers in age(Almost normal)

read_file=read_file[read_file['Age']<(read_file['Age'].mean()+3*read_file['Age'].std())]
read_file.shape
```

```python
#handling outliers from Fare column
#findeng quartiles
Q1=np.percentile(read_file['Fare'],25)
Q3=np.percentile(read_file['Fare'],75)

outlier_low=Q1-1.5*(Q3-Q1)
outlier_high=Q3+1.5*(Q3-Q1)

read_file=read_file[(read_file['Fare']>outlier_low) & (read_file['Fare']<outlier_high)]
```

```python
#one hot encoding#cols to be transfromed are Pclass,Sex,Embarked,family_type
read_file=pd.get_dummies(data=read_file,columns=['Pclass','Sex','Embarked','family_type'],drop_first=True)
```

```python
plt.figure(figsize=(15,6))
sns.heatmap(read_file.corr(),annot=True,cmap='summer')
```

```python
X=read_file.drop("Survived",axis=1)
y=read_file["Survived"]
y.shape
```

```python
def correlation(dataset,threshold):
  col_corr=set()
  corr_matrix=dataset.corr()
  for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i,j]>threshold):
          colname=corr_matrix.columns[i]
          col_corr.add(colname)
  return col_corr
```

```python
X_new=read_file.drop(columns=["Name","Ticket"],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_new_train,X_new_test,y_train,y_test=train_test_split(X_new,y,test_size=.20,random_state=1)
```

```python
X_new.head()
```

```python
#Knearest neighbour
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_new_train,y_train)
```

```python
predicted_knn=knn.predict(X_new_test)
predicted_knn
```

```python
import sklearn.metrics as metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
cm_knn=metrics.confusion_matrix(y_test,predicted_knn)
cmd=ConfusionMatrixDisplay(cm_knn,display_labels=['1','0'])
cmd.plot()
```

```python
from sklearn.metrics import roc_curve
knn_probs=knn.predict_proba(X_new_test)
knn_probs=knn_probs[:,1]
knn_fpr,knn_tpr,_=roc_curve(y_test,knn_probs)
plt.plot(knn_fpr,knn_tpr,marker='.',label=['1','0'])
plt.xlabel('false Positive only')
plt.ylabel('True Positive only')
plt.show()
```

```python
accuracy_knn=metrics.accuracy_score(y_test,predicted_knn)
accuracy_knn
```

```python
#Naive Bayes
from sklearn.naive_bayes import GaussianNB
naive=GaussianNB()
naive.fit(X_new_train,y_train)
```

```python
predicted_naive=naive.predict(X_new_test)
predicted_naive
```

```python
cm_naive=metrics.confusion_matrix(y_test,predicted_naive)
cmd=ConfusionMatrixDisplay(cm_naive,display_labels=['1','0'])
cmd.plot()
```

```python
naive_probs=naive.predict_proba(X_new_test)
naive_probs=naive_probs[:,1]
naive_fpr,naive_tpr,_=roc_curve(y_test,naive_probs)
plt.plot(naive_fpr,naive_tpr,marker='.',label='Naive bayes')
plt.xlabel('False positive only')
plt.ylabel('True positive only')
plt.legend()
plt.show()
```

```python
accuracy_naive=metrics.accuracy_score(y_test,predicted_naive)
accuracy_naive
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_new_train,y_train)
```

```python
predicted_lr=lr.predict(X_new_test)
```

```python
cm_lr=metrics.confusion_matrix(y_test,predicted_lr)
cmd=ConfusionMatrixDisplay(cm_lr,display_labels=['1','0'])
cmd.plot()
```

```python
lr_probs=lr.predict_proba(X_new_test)
lr_probs=lr_probs[:,1]
lr_fpr,lr_tpr,_=roc_curve(y_test,lr_probs)
plt.plot(lr_fpr,lr_tpr,marker='.',label=['1','0'])
plt.xlabel('False Positive Only')
plt.ylabel('True Positive Only')
plt.legend()
plt.show()
```

```python
accuracy_lr=metrics.accuracy_score(y_test,predicted_lr)
accuracy_lr
```
